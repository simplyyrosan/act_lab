```python
d = {
    "Name" : ["Naman", "Anjali", "Diya", "Kuc"],
    "Marks" : [88, 75, 43, 99]
}

df = pd.DataFrame(d)
print(df)
```

```python
print(df.info)
```

```python
print(df.describe)
```

```python
# TO change single value df.loc[row, column] = value

df.loc[2, "Marks"] = 35

print(df)
```

```python
df.iloc[1, 1] = 85

print(df)
```
