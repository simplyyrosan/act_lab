```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5] #Time
y = [20, 30, 35, 12, 40] #Sales

plt.plot(x, y, label = "Sales")
plt.xlabel("Time")
plt.ylabel("Sales")
plt.title("Time vs Sale")
plt.legend()
plt.grid()
plt.show()
```

```python
items = ["A", "B", "C", "D"]
price = [10, 24, 15, 7]

plt.bar(items, price)
plt.xlabel("Items")
plt.ylabel("Price")
plt.title("Items vs Price")
plt.show()
```

```python
import seaborn as sns

x = [1, 2, 3, 4]
y = [10, 20, 30, 40]
sns.lineplot(x=x, y=y)
plt.title("Line Plot using Seaborn")
plt.show()
```
